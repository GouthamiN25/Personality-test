#Simple Linear Regression:
Performed simple linear regression model between the attributes of public, stress, created a scatterplot.

```r
data <- data.frame(
  Public = c( 5.0, 4.0, 5.0, 8.0, 8.0, 9.0, 6.0, 8.7,6.0),
  Stress = c(8.0, 8.0, 10.0, 8.0, 8.5, 7.0, 2.3,1.0,9.0)
)
```

```r
# Create the scatter plot
plot(data$Public, data$Stress,
	xlab = "Public",
	ylab = "Stress",
	main = "Scatter Plot of Public and Stress")
```

```r
# Load ggplot2 libraries
library(ggplot2)

# Plot histograms for numeric variables
ggplot(n_merg, aes(x = Public)) + geom_histogram(binwidth = 5, fill = "red", color = "black") + labs(title = "Public Distribution")
ggplot(n_merg, aes(x = Travel)) + geom_histogram(binwidth = 10000, fill = "pink", color = "black") + labs(title = "Travel Distribution")
ggplot(n_merg, aes(x = Innerpeace)) + geom_histogram(binwidth = 2, fill = "orange", color = "black") + labs(title = "Innerpeace Distribution")
ggplot(n_merg, aes(x = Stress)) + geom_histogram(binwidth = 1, fill = "blue", color = "black") + labs(title = "Stress Distribution")
```

```r
split = sample.split(data$Public, SplitRatio = 0.7)
trainingset = subset(data, split == TRUE)
testset = subset(data, split == FALSE)

# Fitting Simple Linear Regression to the Training set
lm.r= lm(formula = Stress ~ Public,
		data = trainingset)
coef(lm.r)

# Predicting the Test set results
ypred = predict(lm.r, newdata = testset)
```

```r
# Visualising of training set
ggplot() + geom_point(aes(x = trainingset$Public,
						y = trainingset$Stress), colour = 'red') +
geom_line(aes(x = trainingset$Public,
				y = predict(lm.r, newdata = trainingset)), colour = 'blue') +

ggtitle('Public vs Stress (Training set)') +
xlab('Public') +
ylab('Stress')
```

```r
#Correlation Testing using cor()

Public <- c(5.0, 4.0, 5.0, 8.0, 8.0, 9.0, 6.0, 8.7,6.0)
Stress <- c(8.0, 8.0, 10.0, 8.0, 8.5, 7.0, 2.3,1.0,9.0)

# Calculating
# Correlation coefficient
# Using cor() method
result = cor(Public, Stress, method = "pearson")

# Print the result
cat("Pearson correlation coefficient is:", result)
```

```r
# Correlation Testing using cor.test()

# Taking two numeric
# Vectors with same length
Travel <- c(10.0, 5.0, 7.0, 6.0, 8.0, 6.0,6.0,10.0,6.0)
Holiday <- c(5.5,9.0,5.0,8.0,8.5,7.0,2.3,1.0,9.0)

# Calculating
# Correlation coefficient
# Using cor.test() method
result = cor.test(Travel, Holiday, method = "pearson")

# Print the result
print(result)
```

```r
Travel <- c(10.0, 5.0, 7.0, 6.0, 8.0, 6.0, 6.0, 10.0, 6.0)
Holiday <- c(5.5, 9.0, 5.0, 8.0, 8.5, 7.0, 2.3, 1.0, 9.0)

# Calculate correlation
correlation_result <- cor.test(Travel, Holiday, method = "pearson")

# Print the correlation result
print("Correlation Coefficient:")
print(correlation_result$estimate)

# Print the p-value
print("P-value:")
print(correlation_result$p.value)
```
